```python
# %%writefile resume_helper.ipynb
{
 "cells": [
  {
   "cell_type": "markdown",
   "metadata": {},
   "source": [
    "# Resume Helper\n",
    "# Save and restore training state for GitHub + Colab workflow"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "import json\n",
    "import os\n",
    "from datetime import datetime\n",
    "\n",
    "STATE_FILE = \"training_state.json\"\n",
    "\n",
    "def save_state(step, epoch, loss, output_dir, dataset_info):\n",
    "    \"\"\"Save training state to JSON (commit this to GitHub!)\"\"\"\n",
    "    state = {\n",
    "        \"last_step\": step,\n",
    "        \"last_epoch\": epoch,\n",
    "        \"last_loss\": loss,\n",
    "        \"output_dir\": output_dir,\n",
    "        \"dataset_info\": dataset_info,\n",
    "        \"last_updated\": datetime.now().isoformat(),\n",
    "        \"colab_session\": True\n",
    "    }\n",
    "    \n",
    "    with open(STATE_FILE, \"w\") as f:\n",
    "        json.dump(state, f, indent=2)\n",
    "    \n",
    "    print(f\"💾 State saved: Step {step}, Epoch {epoch}\")\n",
    "    return state\n",
    "\n",
    "def load_state():\n",
    "    \"\"\"Load training state from JSON\"\"\"\n",
    "    if not os.path.exists(STATE_FILE):\n",
    "        print(\"No state file found. Starting fresh.\")\n",
    "        return None\n",
    "    \n",
    "    with open(STATE_FILE, \"r\") as f:\n",
    "        state = json.load(f)\n",
    "    \n",
    "    print(f\"📂 Loaded state from {state['last_updated']}\")\n",
    "    print(f\"   Last step: {state['last_step']}\")\n",
    "    print(f\"   Last epoch: {state['last_epoch']}\")\n",
    "    return state\n",
    "\n",
    "def find_latest_checkpoint(output_dir):\n",
    "    \"\"\"Find the most recent checkpoint\"\"\"\n",
    "    if not os.path.exists(output_dir):\n",
    "        return None\n",
    "    \n",
    "    checkpoints = []\n",
    "    for item in os.listdir(output_dir):\n",
    "        if item.startswith(\"checkpoint-\"):\n",
    "            step = int(item.split(\"-\")[-1])\n",
    "            checkpoints.append((step, os.path.join(output_dir, item)))\n",
    "    \n",
    "    if not checkpoints:\n",
    "        return None\n",
    "    \n",
    "    # Return highest step\n",
    "    latest = max(checkpoints, key=lambda x: x[0])\n",
    "    print(f\"🔍 Latest checkpoint: {latest[1]} (Step {latest[0]})\")\n",
    "    return latest[1]\n",
    "\n",
    "print(\"✅ Resume helper loaded!\")"
   ]
  },
  {
   "cell_type": "code",
   "execution_count": null,
   "metadata": {},
   "outputs": [],
   "source": [
    "# QUICK CHECK: Show current status\n",
    "state = load_state()\n",
    "if state:\n",
    "    checkpoint = find_latest_checkpoint(state['output_dir'])\n",
    "    if checkpoint:\n",
    "        print(f\"\\n🚀 Ready to resume from: {checkpoint}\")\n",
    "    else:\n",
    "        print(\"\\n⚠️  State exists but no checkpoint found. May need to restart.\")"
   ]
  }
 ],
 "metadata": {
  "kernelspec": {
   "display_name": "Python 3",
   "language": "python",
   "name": "python3"
  }
 },
 "nbformat": 4,
 "nbformat_minor": 4
}
```
